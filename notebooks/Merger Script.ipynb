```python
import json
import os

DRIVE_BASE = "/content/drive/MyDrive/3rd Year 2nd Term/Machine Learning - CS ELEC 2C /ML - Repo 5/Finals Project/taglish-project"

notebooks = [
    f"{DRIVE_BASE}/01_data_and_eda.ipynb",
    f"{DRIVE_BASE}/02_naive_bayes.ipynb",
    f"{DRIVE_BASE}/03_xlmr_finetune.ipynb",
    f"{DRIVE_BASE}/04_comparison.ipynb",
]

SETUP_KEYWORDS = ["drive.mount", "DRIVE_BASE", "pip install"]

all_cells = []

NOTEBOOK_TITLES = [
    "NOTEBOOK 01 — Data & EDA",
    "NOTEBOOK 02 — Naive Bayes",
    "NOTEBOOK 03 — XLM-RoBERTa Fine-Tuning",
    "NOTEBOOK 04 — Comparison Analysis",
]

for i, nb_path in enumerate(notebooks):
    with open(nb_path, 'r', encoding='utf-8') as f:
        nb = json.load(f)

    # Section divider
    divider = {
        "cell_type": "markdown",
        "metadata": {},
        "source": [
            f"---\n"
            f"# {NOTEBOOK_TITLES[i]}\n"
            f"---"
        ]
    }
    all_cells.append(divider)

    for cell in nb['cells']:
        if i > 0:
            src = "".join(cell.get('source', []))
            if any(kw in src for kw in SETUP_KEYWORDS):
                continue
        all_cells.append(cell)

    print(f"✓ Notebook {i+1} merged — {len(nb['cells'])} cells")

# Use first notebook's metadata as base
with open(notebooks[0], 'r', encoding='utf-8') as f:
    base_nb = json.load(f)

base_nb['cells'] = all_cells

# Add title cell at the very top
title_cell = {
    "cell_type": "markdown",
    "metadata": {},
    "source": [
        "# Sentiment Analysis of Taglish E-Commerce Reviews using Naive Bayes and XLM-RoBERTa\n",
        "**Section:** 3CSE | **Group:** 5\n\n",
        "**Members and Contributions:**\n\n",
        "| Name | Role | Contribution |\n",
        "|------|------|--------------|\n",
        "| [Member 1] | Data & EDA Lead | Notebook 01, Dataset section |\n",
        "| [Member 2] | Classical Model Lead | Notebook 02, NB sections |\n",
        "| [Member 3] (You) | XLM-R Person + Manuscript Lead | Notebook 03, Notebook 04, Manuscript |\n",
        "| [Member 4] | [Role] | [Contribution] |\n",
    ]
}
base_nb['cells'].insert(0, title_cell)

# Save to Drive
output_path = f"{DRIVE_BASE}/3CSE_Group5_Implementation.ipynb"
with open(output_path, 'w', encoding='utf-8') as f:
    json.dump(base_nb, f, indent=1)

print(f"\n✓ Merged notebook saved to Drive.")
print(f"  Path: {output_path}")
print(f"  Total cells: {len(all_cells) + 1}")
```

```python
# Quick sanity check on the merged file
with open(output_path, 'r', encoding='utf-8') as f:
    check = json.load(f)

print(f"Total cells in merged notebook: {len(check['cells'])}")
print(f"\nFirst 5 cell types:")
for cell in check['cells'][:5]:
    src_preview = "".join(cell['source'])[:60].replace('\n', ' ')
    print(f"  [{cell['cell_type']:8s}] {src_preview}")

print(f"\nSection dividers found:")
for cell in check['cells']:
    if cell['cell_type'] == 'markdown':
        src = "".join(cell['source'])
        if 'NOTEBOOK 0' in src:
            print(f"  {src[:50]}")
```
